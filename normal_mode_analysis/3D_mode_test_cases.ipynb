## Cube with square faces vs all vertices connected

```python
model = '3D_cb'
verts, faces, hess, w, v = nma_test_model(model)
verts2, faces2, hess2, w2, v2 = nma_test_model(model = model, verts=None, faces=None, fully_connect=True)
```

```python
sb.distplot(w, kde=False, bins = np.linspace(-0.5, 8.5, 10), label='Edge Connectivity')
sb.distplot(w2, kde=False, bins = np.linspace(-0.5, 8.5, 10), label='Full Connectivity')
plt.xlabel(r'$Energy*m/k$')
plt.ylabel('Number of Modes')
plt.legend();
plt.savefig('cube_hist')
plt.savefig('cube_hist.pdf', format='pdf')
```

## Icosahedron

```python
model = '3D_ico'
```

```python
verts = model_verts[model]
```

```python
x = [verts[i][0] for i in range(verts.shape[0])]
y = [verts[i][1] for i in range(verts.shape[0])]
z = [verts[i][2] for i in range(verts.shape[0])]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x,y,z, color='k')
plt.savefig('icosphere')
plt.savefig('icosphere.pdf', format='pdf')
plt.show()
```

```python
verts, faces, hess, w, v = nma_test_model(model)
```

```python
sb.distplot(w, kde=False, bins = np.linspace(-0.5, 3.5, 12));
plt.xlabel(r'$\omega^2*m/k$')
plt.ylabel('Number of Modes')
plt.savefig('ico_hist')
plt.savefig('ico_hist.pdf', format='pdf')
```

## Larger 3D trimesh shells

```python
verts, faces = trimesh_3D_surface(50, 40, True)
plt.savefig('15verts')
plt.savefig('15verts.pdf', format='pdf')
```

```python
verts, faces, hess, w, v = nma_test_model(model=None, verts=verts, faces=faces, fully_connect=False)
verts2, faces2, hess2, w2, v2 = nma_test_model(model=None, verts=verts, faces=faces, fully_connect=True)
```

```python
sb.distplot(w, kde=False, bins = np.linspace(-0.5, 4.5, 15));
print('Number of modes with w^2 < 10^-5: '+str(sum(w<10**-5)))
plt.xlabel(r'$Energy*m/k$')
plt.ylabel('Number of Modes')
plt.savefig('15verts_hist')
plt.savefig('15verts_hist.pdf', format='pdf')
```

```python
sb.distplot(w2, kde=False, bins = np.linspace(-0.5, 15.5, 16));
plt.xlabel(r'$Energy*m/k$')
plt.ylabel('Number of Modes')
plt.savefig('15verts_fc_hist')
plt.savefig('15verts_fc_hist.pdf', format='pdf')
```
